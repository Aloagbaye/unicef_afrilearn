```python
data = ec[['userId','courseId']]
```

```python
data
```

```python
df = pd.get_dummies(data.set_index('userId')['courseId'].astype(str)).max(level=0).sort_index()
```

```python
df
```

```python
df = df.drop(['nan'],axis=1)
```

```python
#df = df.applymap(lambda x: 1 if x > 0 else 0)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.0001)
#As a data we used frequent_itemsets this time. The one that we created recently above.
rules.sort_values("support", ascending=False).head()
```

```python
sorted_rules = rules.sort_values("lift", ascending=False)
```

```python
enumerate(sorted_rules["antecedents"])
```

```python
def recommend_course(course):
    recommendation_list = []
    for i,product in enumerate(sorted_rules["antecedents"]):
        for j in list(product):
            if j == course:
                recommendation_list.append(list(sorted_rules.iloc[i]["consequents"])[0])
    return recommendation_list
```
